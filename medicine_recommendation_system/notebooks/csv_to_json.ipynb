## Description

* This notebook converts cleaned csv file into json file.

```python
# Import libraries
import pandas as pd
import os
import json
```

```python
# Connect to google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read and copy the data
data = pd.read_csv('YOUR_PATH')
df = data.copy()
```

```python
# Convert pandas DataFrame into a list of dictionaries in JSON format
data_list = json.loads(df.to_json(orient = 'records'))
```

```python
# Writes the JSON data to a file
local_file_path = 'YOUR_PATH'

with open(local_file_path, 'w') as f:
    json.dump(data_list, f)

print(f"File saved at {local_file_path}")
```
